```python
import pandas as pd
import numpy as np

titanic = pd.read_csv('train_and_test2.csv')
titanic
```

```python
titanic.info()
```

```python
passengers = titanic[['Passengerid', 'Age', 'Sex', 'Pclass', 'Fare']]
passengers
```

```python
age = passengers['Age']
print(type(age))
age.describe()
age_clean = age.replace([98, 99], np.nan)
```

```python
fare = passengers['Fare']
print(type(fare))
fare.describe()
fare_clean = fare.replace([98, 99], np.nan)
```

```python
sex = passengers['Sex']
print(type(sex))
sex.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
age_clean.hist(bins=50)
plt.xlabel('Number of Age')
plt.ylabel('Number of People')
plt.title('Age of Passengers in Titanic')
plt.show()
```

```python
fare_clean.hist(bins=50)
plt.xlabel('Amount')
plt.ylabel('Number of People')
plt.title('Amount of Fare in Titanic')
plt.show()
```

```python
gender = list(dict(sex).values())
male = gender.count(0)
female = gender.count(1)
plt.pie([gender.count(0), gender.count(1)], labels=['Female', 'Male'])
# plt.show()
```

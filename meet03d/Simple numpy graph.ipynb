# Using Numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.cos(x)
y4 = np.tan(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y1, color='pink')
plt.title('sin(x)')

plt.subplot(2,2,2)
plt.plot(x, y2, color='purple')
plt.title('sin(2x)')

plt.subplot(2,2,3)
plt.plot(x, y3, color='magenta')
plt.title('cos(x)')

plt.subplot(2,2,4)
plt.plot(x, y4, color='lightpink')
plt.title('tan(x)')

plt.tight_layout()
plt.show()
```

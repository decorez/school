```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1, 11)) #range(1,6 )
square = [n**2 for n in x]
print(square)
```

```python
cubes = [n**3 for n in x]
print(cubes)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, square, label='square')
plt.title('Square Number')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, cubes, label='cubes')
plt.title('Cubes Number')
plt.legend()

randoms = list(range(10, 101, 10))
plt.subplot(2, 2, 3)
plt.plot(x, randoms, label='randoms')
plt.title('Random Number')
plt.legend()

plt.tight_layout()
plt.show()
```

## Using Subplot, Fig, Axes

```python
squares = [x**2 for x in range(1, 1001)]

fig, axes = plt.subplots() #plt.plot()
#axes.plot(squares)
x = list(range(1, 1001))
axes.scatter(x, squares, s=5, c=squares, cmap=plt.cm.Reds)
axes.plot(x, squares)
axes.set_title("Square Number".title(), fontsize=60)
axes.set_xlabel('values', fontsize=30)
axes.set_ylabel('square of value', fontsize=30)

plt.show()
```

```python
cubes = [x**3 for x in range(1, 5001)]
fig, axes = plt.subplots()
x = list(range(1, 5001))
axes.plot(x, cubes)
axes.scatter(x, cubes, s=5, c=cubes, cmap=plt.cm.Reds)
plt.show()
```

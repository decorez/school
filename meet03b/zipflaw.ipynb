## Zip'f Law

```python
import matplotlib.pyplot as plt

book_file = 'pg2600.txt'
first_line = 'CHAPTER I\n'
last_line = '*** END OF THE PROJECT GUTENBERG EBOOK WAR AND PEACE ***'
all_punctuation = '~!@#$%^&*()_+{}|":><?'';-=[]\/'


with open(book_file, 'r', encoding='utf=8') as f:
    lines = f.readlines()
    
start = False
unique_words = {}

for line in lines:
    if line == first_line:
        start = True
        
    if line == last_line:
        break
        
    if start:
        words = line.split()
    
        for word in words:
            word = word.lower()
            word = word.strip(all_punctuation)
            if word in unique_words:
                unique_words[word] += 1
            else:
                unique_words[word] = 1
              

            
freqs = list(unique_words.values())            
freqs.sort(reverse=True)
ranks = list(range(1, len(freqs)+1))
#print(ranks)
plt.xscale('log')
plt.yscale('log')

plt.plot(ranks, freqs)

xi, xf = ranks[0], ranks[-1]
yi, yf = freqs[0], freqs[-1]
plt.plot([xi, xf], [yi, yf], color='gray')

plt.show()
```

```python
sorted_unique_words = dict(sorted(unique_words.items(), key = lambda item : item[1], reverse=True));
words = list(sorted_unique_words.keys())[:5]
freqs = list(sorted_unique_words.values())[:5]

#print(sorted_unique_words)
plt.plot(words, freqs, label='Unique Words')
plt.title('Top 5 Unique Words')
plt.legend()
```
